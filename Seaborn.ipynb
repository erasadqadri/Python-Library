# Seaborn:
used for data visualization and exploratory data analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = sns.load_dataset('flights')
a.head()
```

```python
sns.relplot(x='passengers',y='month',data=a)
```

```python
sns.relplot(x='passengers',y='month',hue='year',data=a)
```

```python
b = sns.load_dataset('tips')
b.head()
```

### Line Plot:

```python
sns.relplot(x='time',y='tip',data=b,kind='line')
```

### CatPlot:
for categorical data

```python
sns.relplot(x='day',y='total_bill',data=b)
```

# Multi Plot Grids:

```python
c = sns.load_dataset('iris')
c.head()
```

```python
d = sns.FacetGrid(c, col='species')
d.map(plt.hist,'sepal_length')
```

```python
e = sns.load_dataset('flights')
e.head()
```

```python
f = sns.PairGrid(e)
f.map(plt.scatter)
```

# Plot Aesthetics:

```python
sns.set(style='dark')
f = sns.PairGrid(e)
f.map(plt.scatter)
```

```python
sns.set(style='darkgrid')
f = sns.PairGrid(e)
f.map(plt.scatter)
```

```python
sns.set(style='whitegrid')
f = sns.PairGrid(e)
f.map(plt.scatter)
```

```python
sns.set(style='white', color_codes=True)

sns.boxplot(x='day',y='total_bill',data=b)
```
