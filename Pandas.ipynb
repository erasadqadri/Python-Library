# Vid 8: How do I filter rows of a pandas DataFrame by column value?

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
movies.shape
```

### we'll filter rows where duratoin >= 200

### Long Method

```python
booleans = []

for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)

booleans[:5] #first five elements of boolean
```

```python
len(booleans) #exactly same as length of the movies dataframe
```

```python
#converting booleans to pandas series
is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies[is_long]
```

### Short Method

```python
is_lengthy = movies.duration >= 200
is_lengthy.head()
```

```python
movies[is_lengthy]
```

### Shortest Method

```python
movies[movies.duration >= 200]
```

### Selecting Columns:

```python
movies[movies.duration >= 200]['genre']
```

```python
movies[movies.duration >= 200]['genre']
```

```python
movies.loc[movies.duration >= 200, 'genre']
```

### drop missing values:

```python
# drop row if any of its column has null value
ufo.dropna(how='any').shape #default
```
